### 2.3. Data Analysis

Our predictive model is trying to predict the pitch type that Max Scherzer will throw based on the above features, representing the current situation of the game and the statistics/performance of the hitter that he is facing. We will use a feature selection technique in our data wrangling process to narrow down the number of features from above to reduce concerns of overfitting. This is a supervised ML problem because we are working with labeled data, and the sub-category of the learning task is classification because we are trying to predict a discrete class label representing the pitch type that Max Scherzer throws.

We are going to use the following four different machine learning algorithms: 
* **Naive Bayes:** This is a probabilistic classification algorithm. It assumes that the features are equally important and statistically independent. It works by finding the probability that the given sample belongs to each class and assigns the sample to the class with the highest probability. We are using it because it tends to work well with high-dimensional data and also as a baseline comparison against our other models.

* **Decision Tree:** This classification algorithm creates a tree of if-else questions to answer based on the features to classify each sample. We are using this algorithm as a simple model to use for a baseline comparison against our other models.

* **k-Nearest Neighbors:** This classification algorithm works by looking at the k samples nearest in distance from the one you are trying to classify and classifies the sample based on the class with the most samples in the k nearest. We are using this algorithm because we think that it will perform well as it will look at the most similar situations and predict the pitch type that Max Scherzer has historically thrown in the past in similar situations.

* **Support Vector Machines:** This classification algorithm works by finding and using the hyper-plane that differentiates the classes very well and that maximizes this margin between the hyper-plane and the nearest data points. We are using this algorithm because we think it will work well as it tends to work well with large datasets like ours, and it is easy and simple to train.

**Changing the Target Value to a Numeric Value**

Our dataset uses a string to represent the type of pitch thrown, but our machine learning algorithms need all values to be numeric, so we decided to use a dictionary to assign each of the five different pitch types to a value one through five. The value of fastball is 1, change-up is 2, curveball is 3, slider is 4, and cutter is 5.

```python
# defines dictionary to change pitch type to numeric values
pitch_thrown = {"FF":int(1), "CH":int(2), "CU":int(3), "SL":int(4), "FC":int(5)}

# replaces string values with the corresponding numeric value
df_pitches = df_pitches.replace(pitch_thrown)

# sample output
df_pitches
```

**Changing Features to Make Them All Numeric**

In our dataset, many of our features were of type string because they had a percent sign next to the numeric value when we scraped the data from the websites. As mentioned above, our machine learning algorithms need all values to be numeric in the dataset, so we defined a function to change all of the percentages of type strings to numeric decimal values, so that they are compatible with our machine learning algorithms.

```python
# returns a numeric decimal value from a string that represents a percent
def percent_changer(x):
    return float(x.strip('%'))/100

# maps the percent-changer function to all columns that are string percent values
df_pitches["BB%"] = df_pitches["BB%"].map(percent_changer)
df_pitches["K%"] = df_pitches["K%"].map(percent_changer)
df_pitches["LD%"] = df_pitches["LD%"].map(percent_changer)
df_pitches["GB%"] = df_pitches["GB%"].map(percent_changer)
df_pitches["FB%"] = df_pitches["FB%"].map(percent_changer)
df_pitches["FA% (pfx)"] = df_pitches["FA% (pfx)"].map(percent_changer)
df_pitches["FC% (pfx)"] = df_pitches["FC% (pfx)"].map(percent_changer)
df_pitches["SL% (pfx)"] = df_pitches["SL% (pfx)"].map(percent_changer)
df_pitches["CU% (pfx)"] = df_pitches["CU% (pfx)"].map(percent_changer)
df_pitches["CH% (pfx)"] = df_pitches["CH% (pfx)"].map(percent_changer)
df_pitches["O-Swing% (pfx)"] = df_pitches["O-Swing% (pfx)"].map(percent_changer)
df_pitches["Z-Swing% (pfx)"] = df_pitches["Z-Swing% (pfx)"].map(percent_changer)
df_pitches["Swing% (pfx)"] = df_pitches["Swing% (pfx)"].map(percent_changer)
df_pitches["O-Contact% (pfx)"] = df_pitches["O-Contact% (pfx)"].map(percent_changer)
df_pitches["Z-Contact% (pfx)"] = df_pitches["Z-Contact% (pfx)"].map(percent_changer)
df_pitches["Contact% (pfx)"] = df_pitches["Contact% (pfx)"].map(percent_changer)
df_pitches["Pull%"] = df_pitches["Pull%"].map(percent_changer)
df_pitches["Cent%"] = df_pitches["Cent%"].map(percent_changer)
df_pitches["Oppo%"] = df_pitches["Oppo%"].map(percent_changer)

# sample output
df_pitches
```

**Dropping Outlier Pitch Types**

In our dataset, there were some pitches that were not one of the five pitch types that we are looking at predicting. These pitches were rare as they are not ones that Max Scherzer throw often. We thought that it would be the best idea to drop the rows of these pitches because we consider them outliers that do not happen very often, and we think it would help out our models' prediction accuracy if we drop them. We did this by iterating through each row of the dataframe and dropping the row if the "pitch_thrown" value was not one of the five main pitches we are looking for.

We should mention that we chose which pitches to consider by referring to which pitches are registered on http://www.brooksbaseball.net/. This website has comprehensive pitch data for every pitcher in Major League Baseball, and we figured that any pitches not registered for Scherzer (http://www.brooksbaseball.net/landing.php?player=453286) were probably not thrown enough to be considered one that is in his repertoire.

```python
# prints current amount of pitches
print(len(df_pitches))

# for each row in the dataset
for index, row in df_pitches.iterrows():
    # if the pitch_thrown value does not equal 1 through 5
    if row["pitch_thrown"] != 1.0 and row["pitch_thrown"] != 2.0 and row["pitch_thrown"] != 3.0 and row["pitch_thrown"] != 4.0 and row["pitch_thrown"] != 5.0:
        # drop the row
        df_pitches.drop(index, inplace=True)

# prints amount of pitches after dropping outlier pitch types
print(len(df_pitches))

# resets the index values
df_pitches.reset_index(inplace = True)
df_pitches.drop(["index"], axis = 1, inplace = True)

# sample output
df_pitches
```

We decided to use One-Hot Encoding as our technique to change these values. **Scaling with MinMaxScaler**

Next, we decided to preprocess our data by scaling it using the scikit-learn function, MinMaxScaler. MinMaxScaler transforms all of the numeric feature values so that that the data is normalized to a scale ranging between 0 and 1. We are scaling our data because we think that it will lead to our models having better results because it will bring all of our data onto the same scale instead of the varying magnitude and units that our data started with, especially since some of the algorithms rely on euclidean distance.

```python
# scale data using MinMaxScaler
def preprocessing(X_train, X_test):
    
    #instantiate
    scaler = MinMaxScaler()
    
    # fit the scaler to the training set
    scaler.fit(X_train)
    
    # scale testing and training sets
    X_train_scaled = scaler.transform(X_train)
    X_test_scaled = scaler.transform(X_test)

    # return scaled training and testing sets
    return X_train_scaled, X_test_scaled
```

### Feature Selection

**Recursive Feature Elimination**

Since our current dataset has 41 different features, we have concerns of overfitting our model because with too high a number of features, the model can become too complex, leading to lower performance and accuracy. To ease these concerns, we decided to use recursive feature elimination to select a subset of our original features to use to train and test our model. RFE iteratively builds a model and eliminates the least important feature to the model each time until the desired number of features is left, which in our case is 5. We decided to select 5 because we did not want to make our models too complex, which would lead to overfitting, and we played around with our models and the number of features and found that 5 features lead to optimized results.

```python
# selects the 5 most important features to be used for our model
def feature_selection(x_train, x_test, y_train, features):
    
    # instantiate
    select = RFE(DecisionTreeRegressor(random_state = 3000), n_features_to_select = 5)
    
    # fit the RFE selector to the training data
    select.fit(x_train, y_train)
    
    # transform training and testing sets so only the selected features are retained
    X_train_selected = select.transform(x_train)
    X_test_selected = select.transform(x_test)
    
    # prints selected features/Sample Output
    selected_features = [feature for feature, status in zip(features, select.get_support()) if status == True]
    print('Selected features:')
    for feature in selected_features:
        print(feature)

    # returns selected features
    return X_train_selected, X_test_selected
```

```python
# helps us get the distribution of every pitch based on a time-based condition (inning, pitch number, etc.)
# there is a cut-off at 10 because it provides the cumulative distribution of extra-inning appearances (which did not happen
# between 2015 and 2019 for Max Scherzer) or pitches that were the 10th or further pitch in a given at-bat (which is extremely
# rare).

def distrib_finder(num, col):
    FF = 0
    CH = 0
    CU = 0
    SL = 0
    FC = 0
    TOTAL = 0
    
    row = 0
    
    # checks if the value is >= 10, meaning the distribution is cumulative for pitches with num >= 10
    if num >= 10:
        
        # iterate through the rows of df_pitches
        while row < len(df_pitches):
            
            # check if the pitch is to-be accounted for in the distribution
            if df_pitches.iloc[row][col] >= 10:
                
                # add the pitch to the pitch count and total count
                if df_pitches.iloc[row]["pitch_thrown"] == 1.0:
                    FF += 1
                    TOTAL += 1
                elif df_pitches.iloc[row]["pitch_thrown"] == 2.0:
                    CH += 1
                    TOTAL += 1
                elif df_pitches.iloc[row]["pitch_thrown"] == 3.0:
                    CU += 1
                    TOTAL += 1
                elif df_pitches.iloc[row]["pitch_thrown"] == 4.0:
                    SL += 1
                    TOTAL += 1
                elif df_pitches.iloc[row]["pitch_thrown"] == 5.0:
                    FC += 1
                    TOTAL += 1
                    
            # iterate through the rows
            row = row + 1
    
    # works for num values below 10
    else:
        while row < len(df_pitches):
            if df_pitches.iloc[row][col] == num:
                if df_pitches.iloc[row]["pitch_thrown"] == 1.0:
                    FF += 1
                    TOTAL += 1
                elif df_pitches.iloc[row]["pitch_thrown"] == 2.0:
                    CH += 1
                    TOTAL += 1
                elif df_pitches.iloc[row]["pitch_thrown"] == 3.0:
                    CU += 1
                    TOTAL += 1
                elif df_pitches.iloc[row]["pitch_thrown"] == 4.0:
                    SL += 1
                    TOTAL += 1
                elif df_pitches.iloc[row]["pitch_thrown"] == 5.0:
                    FC += 1
                    TOTAL += 1
            row = row + 1
        
    # calculate percentages
    FF_percent = (FF / TOTAL) * 100
    CH_percent = (CH / TOTAL) * 100
    CU_percent = (CU / TOTAL) * 100
    SL_percent = (SL / TOTAL) * 100
    FC_percent = (FC / TOTAL) * 100
        
    # return values
    return FF_percent, CH_percent, CU_percent, SL_percent, FC_percent
```

```python
# create methods for each algorithm to predict the pitch type
# return 0.0 is the prediction is wrong or 1.0 if the prediction is right
def predict_knn(val0, val1, val2, val3, val4, target):    
    esti = KNeighborsClassifier(metric = 'euclidean', n_neighbors = 9)
    esti.fit(X = X_train_selected, y = y_train)
    x_val = np.array([val0, val1, val2, val3, val4]).reshape(1, -1)    
    predicted = esti.predict(X = x_val)    
    accuracy = esti.score(x_val, [target])
    return accuracy

def predict_linear(val0, val1, val2, val3, val4, target):    
    esti = LinearSVC(random_state = 3000, max_iter = 1000000, C = 0.01)
    esti.fit(X = X_train_selected, y = y_train)
    x_val = np.array([val0, val1, val2, val3, val4]).reshape(1, -1)    
    predicted = esti.predict(X = x_val)    
    accuracy = esti.score(x_val, [target])
    return accuracy

def predict_tree(val0, val1, val2, val3, val4, target):    
    esti = DecisionTreeClassifier(max_depth = 100, min_samples_split = 1000)
    esti.fit(X = X_train_selected, y = y_train)
    x_val = np.array([val0, val1, val2, val3, val4]).reshape(1, -1)    
    predicted = esti.predict(X = x_val)    
    accuracy = esti.score(x_val, [target])
    return accuracy

def predict_bayes(val0, val1, val2, val3, val4, target):    
    esti = GaussianNB()
    esti.fit(X = X_train_selected, y = y_train)
    x_val = np.array([val0, val1, val2, val3, val4]).reshape(1, -1)    
    predicted = esti.predict(X = x_val)    
    accuracy = esti.score(x_val, [target])
    return accuracy

# create method to run a prediction based on features, a target, and a provided algorithm
def predict(val0, val1, val2, val3, val4, target, test):
    if test == "KNeighborsClassifier":
        return predict_knn(val0, val1, val2, val3, val4, target)
    if test == "LinearSVC":
        return predict_linear(val0, val1, val2, val3, val4, target)
    if test == "DecisionTreeClassifier":
        return predict_tree(val0, val1, val2, val3, val4, target)
    if test == "GaussianNaiveBayes":
        return predict_bayes(val0, val1, val2, val3, val4, target)
```
